## Twitter API

```python
CONSUMER_KEY = ''
CONSUMER_SECRET_KEY = ''

ACCESS_TOKEN = ''
ACCESS_SECRET_TOKEN = ''
```

```python
AUTH = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET_KEY)
AUTH.set_access_token(ACCESS_TOKEN, ACCESS_SECRET_TOKEN)

# Initialize API
API = tweepy.API(AUTH)
```

## Getting the Tweet

```python
QUERY = 'ppkm OR vaksin OR lockdown OR covid OR corona OR pandemi' # Ganti query
TWEETS = 900 # Tweets per run. [Kalo gasalah 900 tweets maximum per request.]
RUNS = 1 # Run yang mau dijalanin.
```

```python
df_tweets = pd.DataFrame(columns = ['text'])

p_start = time.time()
for i in range(0, RUNS):
  start_run = time.time()

  tweet_list = tweepy.Cursor(API.search, q=QUERY + ' -filter:retweets',
                              lang="id", since="2021-07-19", until="2021-07-20",
                              tweet_mode='extended').items(TWEETS)

  num_tweets_added = 0
  for tweet in tweet_list:

    # Adding elements
    # created = tweet.created_at
    # username = tweet.user.screen_name
    # location = tweet.user.location
    # retweet =  tweet.retweet_count

    try:
      text = tweet.retweeted_status.full_text
    except:
      text = tweet.full_text
  
    n_tweets = [text]
    df_tweets.loc[len(df_tweets)] = n_tweets

    num_tweets_added += 1

  end_run = time.time()
  interval = round((end_run-start_run)/60, 2)

  print('RUN {}'.format(i+1))
  print('=====')
  print('Num Tweets Scraped {}'.format(num_tweets_added))
  print('Time {}'.format(interval))

  if i+1 != RUNS:
    time.sleep(920) # Kalo lebih dari sekali run uncomment ini.
```

```python
# Obtain timestamp in a readable format
TIMESTAMP = datetime.date.today().strftime('%Y%m%d_%H%M%S')

# Define working path and filename
PATH = os.getcwd()
FILENAME = PATH + '/' + TIMESTAMP + '.csv'

# Store dataframe in csv with creation date timestamp
df_tweets.to_csv(FILENAME, index = False)
    
p_end = time.time()

print('Scraping has completed!')
print('Total time taken to scrap is {} minutes.'.format(round(p_end-p_start)/60, 2))
```
